```python
import matplotlib.pyplot as plt
from util import finance
from util import ts
```

```python
def plot_spectral_entropy(symbol: str):
    df = finance.get_stock_data(symbol, '2010-01-01', '2023-12-31')
    fig, axs = plt.subplots(4, 1, figsize=(12, 8))
    fig.subplots_adjust(hspace=.5)
    fig.suptitle(symbol, fontsize=12)
    axs[0].plot(df['Close'], label=f'SE={ts.spectral_entropy(df["Close"].values):.2f}')
    axs[0].set_title('Close')
    axs[0].legend()
    axs[1].plot(df['Return'], label=f'SE={ts.spectral_entropy(df["Return"].values):.2f}')
    axs[1].set_title(f'Return')
    axs[1].legend()
    axs[2].plot(df['Log_Close'], label=f'SE={ts.spectral_entropy(df["Log_Close"].values):.2f}')
    axs[2].set_title('Log Close')
    axs[2].legend()
    axs[3].plot(df['Log_Return'], label=f'SE={ts.spectral_entropy(df["Log_Return"].values):.2f}')
    axs[3].set_title('Log Return')
    axs[3].legend()
    plt.show()
```

```python
plot_spectral_entropy('^GSPC')
plot_spectral_entropy('AAPL')
plot_spectral_entropy('HUF=X')
plot_spectral_entropy('EURUSD=X')
plot_spectral_entropy('GC=F')
plot_spectral_entropy('CL=F')
```
